```python
df = pd.read_csv("CSV_Files/data 5.csv")
```

```python
df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df.sort_values('Date',inplace=True)
```

```python
dates = df['Date']
price_open = df['Open']
price_close = df['Close']
high = df['High']
low = df['Low']
```

```python
plt.plot(dates,
        price_close,
        marker='o',
        label='Closing Price')
plt.plot(dates,
        price_open,
        marker='o',
        label='Opening Price')
plt.plot(dates,
        high,linestyle='--',
        marker='o',
        label = 'High')
plt.plot(dates,
        low,linestyle='--',
        marker='o',
        label = 'Low')

plt.legend()
plt.gcf().autofmt_xdate()
plt.title("BitCoin Prices")
plt.xlabel("Date")
plt.ylabel("Closing Price")
plt.tight_layout()
```
