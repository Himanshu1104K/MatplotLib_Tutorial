```python
x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]
```

```python
plt.style.available
```

```python
plt.style.use("seaborn-v0_8-colorblind")
plt.scatter(x,y)
plt.title("My First Scatter Plot")
plt.xlabel("X-values")
plt.ylabel("Y-values")
plt.tight_layout()
```

```python
plt.style.use("seaborn-v0_8-colorblind")
plt.scatter(x,y,s=100,c='green')
plt.title("My First Scatter Plot")
plt.xlabel("X-values")
plt.ylabel("Y-values")
plt.tight_layout()
```

```python
plt.style.use("seaborn-v0_8-colorblind")

plt.scatter(x,y,
            s=100,
            c='green',
            linewidths=1,
            edgecolors='black',
            alpha=0.75)

plt.title("My First Scatter Plot")
plt.xlabel("X-values")
plt.ylabel("Y-values")
plt.tight_layout()
```

```python
df = pd.read_csv("CSV_Files/2019-05-31-data.csv")
```

```python
df.head()
```

```python
view_count = df['view_count']
likes = df['likes']
ratio = df['ratio']
```

```python
plt.style.use("seaborn-v0_8-colorblind")
plt.scatter(view_count,likes)
plt.title("Trending Youtube Videos")
plt.xlabel('View Counts')
plt.ylabel("Number of Likes")
```

```python
plt.style.use("seaborn-v0_8-colorblind")
plt.scatter(view_count,likes,
            c=ratio,
            cmap='Greens',
            edgecolors='black')

cbar = plt.colorbar()
cbar.set_label("Likes to Dislike Ratio")

plt.title("Trending Youtube Videos")
plt.xlabel('View Counts')
plt.ylabel("Number of Likes")
plt.tight_layout()
```

```python
plt.style.use("seaborn-v0_8-colorblind")
plt.scatter(view_count,likes,
            c=ratio,
            cmap='Greens',
            edgecolors='black')

plt.xscale('log')
plt.yscale('log')

cbar = plt.colorbar()
cbar.set_label("Likes to Dislike Ratio")

plt.title("Trending Youtube Videos")
plt.xlabel('View Counts')
plt.ylabel("Number of Likes")
plt.tight_layout()
```

```python
plt.style.use("seaborn-v0_8-colorblind")
plt.scatter(view_count,likes,
            c=ratio,
            cmap='summer',
            edgecolors='black')

plt.xscale('log')
plt.yscale('log')

cbar = plt.colorbar()
cbar.set_label("Likes to Dislike Ratio")

plt.title("Trending Youtube Videos")
plt.xlabel('View Counts')
plt.ylabel("Number of Likes")
plt.tight_layout()
```
