```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
ages_x = np.arange(25,36)

dev_y = [38496, 42000, 46752, 49320, 53200,
         56000, 62316, 64928, 67317, 68748, 73752]

py_dev_y = [45372, 48876, 53850, 57287, 63016,
            65998, 70003, 70000, 71496, 75370, 83640]

js_dev_y = [37810, 43515, 46823, 49293, 53437,
            56373, 62375, 66674, 68745, 68746, 74583]
```

```python
plt.plot(ages_x,dev_y,color='k',label='All Dev')
plt.plot(ages_x,py_dev_y,color='g',label='Python')
plt.plot(ages_x,js_dev_y,color='b',label='JavaScript')
plt.title("Median Salary (USD) by Age")
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.legend()
plt.tight_layout()
```

In the above cell we overcome the bar chart stacking but due to this our xlabels are all messed up so lets fix it 

```python
x_indexes = np.arange(len(ages_x))
width = 0.25
plt.bar(x_indexes-width,dev_y,width=width,color='k',label='All Dev')
plt.bar(x_indexes,py_dev_y,width=width,color='g',label='Python')
plt.bar(x_indexes+width,js_dev_y,width=width,color='b',label='JavaScript')
plt.title("Median Salary (USD) by Age")
plt.xticks(ticks=x_indexes,labels=ages_x) #this function help us to label the x indexes with lebels as ages of x
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.legend()
plt.tight_layout()
```

```python
import csv
from collections import Counter
```

```python
counter = Counter()
with open('CSV_Files/data.csv') as csv_file:
    csv_reader = csv.DictReader(csv_file)
    for row in csv_reader:
        counter.update(row['LanguagesWorkedWith'].split(";"))
```

```python
counter.most_common(15)
```

```python
for x,y in counter.most_common(15):
    plt.bar(x,y,color='b')
plt.title("Most Popular Languages")
plt.ylabel("No. of People Who Use")
plt.xlabel("Programming Languages")
plt.tight_layout()
```

```python
for x,y in counter.most_common(15):
    plt.barh(x,y,color='b')
plt.title("Most Popular Languages",color='green')
plt.xlabel("No. of People Who Use")
plt.tight_layout()
```
