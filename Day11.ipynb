```python
from matplotlib import pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('CSV_Files/data2.csv')
```

```python
ages = df['Age']
all_dev = df['All_Devs']
py_sal = df['Python']
js_sal = df['JavaScript']
```

```python
plt.plot(ages,py_sal,label='Python')
plt.plot(ages,js_sal,label='JavaScript')
plt.plot(ages,all_dev,label='All Developers',color='k',linestyle="--")
plt.xlabel("Ages")
plt.ylabel("Median Salary (USD)")
plt.title("Median Salary (USD) by Age")
plt.legend()
plt.tight_layout()
```

```python
fig , ax = plt.subplots()

ax.plot(ages,py_sal,label='Python')
ax.plot(ages,js_sal,label='JavaScript')
ax.plot(ages,all_dev,label='All Developers',color='k',linestyle="--")
ax.set_xlabel("Ages")
ax.set_ylabel("Median Salary (USD)")
ax.set_title("Median Salary (USD) by Age")
ax.legend()
plt.tight_layout()
```

```python
fig , (ax1,ax2) = plt.subplots(nrows=2,ncols=1)

ax1.plot(ages,all_dev,label='All Developers',color='k',linestyle="--")
ax2.plot(ages,py_sal,label='Python')
ax2.plot(ages,js_sal,label='JavaScript')

ax1.set_xlabel("Ages")
ax1.set_ylabel("Median Salary (USD)")
ax1.set_title("Median Salary (USD) by Age")
ax1.legend()


ax2.set_xlabel("Ages")
ax2.set_ylabel("Median Salary (USD)")
ax2.set_title("Median Salary (USD) by Age")
ax2.legend()
plt.tight_layout()
```
