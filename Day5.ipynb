```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('CSV_Files/data2.csv')
```

```python
df.head()
```

```python
ages = df['Age']
dev_sal = df['All_Devs']
py_sal = df['Python']
js_sal = df['JavaScript']
```

```python
plt.plot(ages,dev_sal,linestyle='--',color='#444444',label='All Devs')
plt.plot(ages,py_sal,label='Python')
# plt.plot(ages,js_sal,label='JavaScript')
plt.tight_layout()
plt.grid()
plt.title("Devloper Salary Based on Ages and Language")
plt.xlabel("Ages")
plt.ylabel("Median Salary (USD)")
plt.legend()
```

```python
plt.plot(ages,dev_sal,linestyle='--',color='#444444',label='All Devs')
plt.plot(ages,py_sal,label='Python')

plt.fill_between(ages,py_sal)

plt.tight_layout()
plt.title("Devloper Salary Based on Ages and Language")
plt.xlabel("Ages")
plt.ylabel("Median Salary (USD)")
plt.legend()
```

```python
plt.plot(ages,dev_sal,linestyle='--',color='#444444',label='All Devs')
plt.plot(ages,py_sal,label='Python')

plt.fill_between(ages,py_sal , alpha=0.25)

plt.tight_layout()
plt.title("Devloper Salary Based on Ages and Language")
plt.xlabel("Ages")
plt.ylabel("Median Salary (USD)")
plt.legend()
```

```python
median_sal = (df[['Python','All_Devs','JavaScript']].median()).median()
median_sal
```

```python
plt.plot(ages,dev_sal,linestyle='--',color='#444444',label='All Devs')
plt.plot(ages,py_sal,label='Python')

plt.fill_between(ages,py_sal , median_sal, alpha=0.25)

plt.title("Devloper Salary Based on Ages and Language")
plt.xlabel("Ages")
plt.ylabel("Median Salary (USD)")
plt.legend()
```

```python
plt.plot(ages,dev_sal,linestyle='--',color='#444444',label='All Devs')
plt.plot(ages,py_sal,label='Python')

plt.fill_between(ages, py_sal,
                median_sal, 
                where=(py_sal>median_sal),
                interpolate=True,alpha=0.25)

plt.title("Devloper Salary Based on Ages and Language")
plt.xlabel("Ages")
plt.ylabel("Median Salary (USD)")
plt.legend()
```

```python
plt.plot(ages,dev_sal,linestyle='--',color='#444444',label='All Devs')
plt.plot(ages,py_sal,label='Python')

plt.fill_between(ages, py_sal,
                median_sal, 
                where=(py_sal>median_sal),
                interpolate=True,alpha=0.25)
plt.fill_between(ages, py_sal,
                median_sal, 
                where=(py_sal<=median_sal),
                interpolate=True,alpha=0.25)

plt.title("Devloper Salary Based on Ages and Language")
plt.xlabel("Ages")
plt.ylabel("Median Salary (USD)")
plt.legend()
```

```python
plt.plot(ages,dev_sal,linestyle='--',color='#444444',label='All Devs')
plt.plot(ages,py_sal,label='Python')

plt.fill_between(ages, py_sal,
                dev_sal, 
                where=(py_sal>dev_sal),
                interpolate=True,alpha=0.25)
plt.fill_between(ages, py_sal,
                dev_sal, 
                where=(py_sal<=dev_sal),
                interpolate=True,alpha=0.25)

plt.title("Devloper Salary Based on Ages and Language")
plt.xlabel("Ages")
plt.ylabel("Median Salary (USD)")
plt.legend()
```

```python
plt.plot(ages,dev_sal,linestyle='--',color='#444444',label='All Devs')
plt.plot(ages,py_sal,label='Python')

plt.fill_between(ages, py_sal,
                dev_sal, 
                where=(py_sal>dev_sal),
                interpolate=True,alpha=0.25,label='Above Avg')
plt.fill_between(ages, py_sal,
                dev_sal, 
                where=(py_sal<=dev_sal),
                interpolate=True,alpha=0.25,label='Below Avg')

plt.title("Devloper Salary Based on Ages and Language")
plt.xlabel("Ages")
plt.ylabel("Median Salary (USD)")
plt.legend()
```
