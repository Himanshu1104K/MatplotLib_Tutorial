```python
import matplotlib.pyplot as plt
```

```python
slices = [120,60,30,10,50,20]
labels = ['Slices1','Slices2','Slices3','Slices4','Slices5','Slices6']
```

```python
plt.pie(slices,labels=labels)
```

```python
plt.pie(slices,labels=labels,wedgeprops={'edgecolor':'white'})
plt.title("My Awesome Pie Chart")
plt.tight_layout()
```

```python
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
```

```python
plt.pie(slices,labels=labels,wedgeprops={'edgecolor':'white'})
plt.title("My Awesome Pie Chart")
plt.tight_layout()
```

```python
explode = [0,0,0,0.1,0]
plt.pie(slices,
        labels=labels,
        explode=explode,
        wedgeprops={'edgecolor':'white'})
plt.title("My Awesome Pie Chart")
plt.tight_layout()
```

```python
plt.pie(slices,
        labels=labels,
        explode=explode,
        shadow=True,
        wedgeprops={'edgecolor':'white'})
plt.title("My Awesome Pie Chart")
plt.tight_layout()
```

```python
plt.pie(slices,
        labels=labels,
        explode=explode,
        shadow=True,
        startangle=90,
        wedgeprops={'edgecolor':'white'})
plt.title("My Awesome Pie Chart")
plt.tight_layout()
```

```python
plt.pie(slices,
        labels=labels,
        explode=explode,
        shadow=True,
        autopct='%1.1f%%',
        startangle=90,
        wedgeprops={'edgecolor':'white'})
plt.title("My Awesome Pie Chart")
plt.tight_layout()
```
